```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1D case

Syntetic dataset

```python
M = 100

x_min = 0.0
x_max = 10.0
x = np.linspace(start=x_min, stop=x_max, num=M)
```

```python
a0_exact = 1.0
a1_exact = 0.5
y_exact = a0_exact + a1_exact * x
plt.plot(x, y_exact)

noise_scale = 2.0
y = y_exact + noise_scale * (np.random.rand(M) * 2 - 1)
plt.plot(x, y)
```

Batch Gradient Descent

```python
a0, a1 = 0, 0   # Inicialization of the parameters (it could be randomized)

lr = 1e-4       # Learning Rate
num_iter = 1000

for k in range(num_iter):

    # Prediction at iteration i
    y_hat = a0 + a1 * x

    grad_0, grad_1 = 0.0, 0.0
    error = 0.0
    square_sum = 0.0
    for i in range(M):
        error_i = y_hat[i] - y[i]

        # Components of the gradient vector
        grad_0 += error_i * 1
        grad_1 += error_i * x[i]

        # Loss function
        square_sum += error_i**2

    loss = (1 / 2) * square_sum

    a0 -= lr * grad_0
    a1 -= lr * grad_1

print(f"Loss = {loss}, y = {a0} + {a1} * x")
```
